```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

runtime = pd.read_csv('runtime.csv',header=None)
columns = ['bdsim', 'pyblksim',]# 'matlab_closed', 'matlab_open']
runtime.columns=columns
runtime['number of simulations'] = run

memory = pd.read_csv('memory.csv',header=None)
memory.columns=columns
memory['number of simulations'] = run
```

```python
runtime2 = pd.read_csv('runtime2.csv',header=None)
runtime2.columns = ['matlab_closed', 'matlab_open']
memory2 = pd.read_csv('memory2.csv',header=None)
memory2.columns = ['matlab_closed', 'matlab_open']
```

```python
runtime.head()
```

```python
df = pd.concat([runtime, runtime2], axis=1)
df1 = pd.concat([memory,memory2],axis=1)
```

```python
plt.figure(figsize=(12,6),dpi=200)
plt.yscale('log')
plt.plot(df['number of simulations'], df['bdsim'], label='bdsim', marker='.')
plt.plot(df['number of simulations'], df['pyblksim'], label='pyblksim', marker='.')
plt.plot(df['number of simulations'], df['matlab_closed'], label='matlab_closed', marker='.')
plt.plot(df['number of simulations'], df['matlab_open'], label='matlab_open', marker='.')


plt.xticks([1,2,3,5,10,15,20,25,30,35,40,45,50])
plt.yticks(ticks=[1,2,3,4,5,10,20,30],fontsize=12)  # Customize y-ticks

# Add grid for better visibility
plt.grid(True, which="both", ls="--", linewidth=0.5)

# Add legend
plt.legend()
plt.xlabel('Number of Simulations')
plt.ylabel('Runtime')
plt.title('Runtime vs Number of Simulations')
```

```python
plt.figure(figsize=(12,6),dpi=200)
plt.yscale('log')
plt.plot(df1['number of simulations'], df1['bdsim'], label='bdsim', marker='.')
plt.plot(df1['number of simulations'], df1['pyblksim'], label='pyblksim', marker='.')
plt.plot(df1['number of simulations'], df1['matlab_closed'], label='matlab_closed', marker='.')
plt.plot(df1['number of simulations'], df1['matlab_open'], label='matlab_open', marker='.')


plt.xticks([1,2,3,5,10,15,20,25,30,35,40,45,50])
#plt.yticks(ticks=[1,2,3,4,5,10,20,30],fontsize=12)  # Customize y-ticks

# Add grid for better visibility
plt.grid(True, which="both", ls="--", linewidth=0.5)

# Add legend
plt.legend()
plt.xlabel('Number of Simulations')
plt.ylabel('Net Memory')
plt.title('Net Memory vs Number of Simulations')
```

```python
df_run = df.drop('number of simulations',axis=1)
df_run.to_csv('new_runtime.csv',index=False)

df_mem = df1.drop('number of simulations',axis=1)
df_mem.to_csv('new_memory.csv',index=False)
```
